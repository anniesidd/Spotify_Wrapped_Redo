```python
import pandas as pd
import numpy as np
```

```python
music1 = pd.read_json('Spotify Account Data/StreamingHistory_music_0.json')
# music1
```

```python
music2 = pd.read_json('Spotify Account Data/StreamingHistory_music_0.json')
# music2
```

```python
music = pd.concat([music1,music2])
music
```

```python
# What is my time period
print("First date: ", music['endTime'].min())
print("Last date: ", music['endTime'].max())
```

```python
# New Column with Track Name and Artist in case of duplicates
music['Track_Artist'] = music['trackName'] + "_" + music['artistName']
```

```python
# See my top songs
song_counts = music[['Track_Artist', 'trackName']].groupby('Track_Artist').count()
song_counts.rename(columns={'trackName': 'Count'}, inplace=True)

top_songs = song_counts.sort_values('Count', ascending=False).head(20)
top_songs
```

```python
# Looking at Christmas Music
december = music[(music['endTime'] > '2024-11-30') & (music['endTime'] < '2025-01-01')]
song_counts_dec = december[['Track_Artist', 'trackName']].groupby('Track_Artist').count()
song_counts_dec.rename(columns={'trackName': 'Count'}, inplace=True)

top_songs_dec = song_counts_dec.sort_values('Count', ascending=False).head(10)
top_songs_dec
```
